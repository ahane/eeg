```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from collections import Counter
import math
```

```python
import pickle
with open('all_pv.df.pick', 'r') as f:
    all_pv = pickle.load(f)
```

```python
def date_to_coord(date_col):
    timedeltas_to_left = - (date_col.min() - date_col)
    days_to_left = (timedeltas_to_left / np.timedelta64(1, 'D'))
    return days_to_left.astype(np.int)
```

```python
def kw_to_coord(kw_col):
    floored = np.floor(kw_col).astype(np.int)
    max_kw = floored.max()
    return max_kw - floored
    
```

```python
import Image
from IPython.core import display
from io import BytesIO

def display_pil_image(im):
    """displayhook function for PIL Images, rendered as PNG"""
    b = BytesIO()
    im.save(b, format='png')
    data = b.getvalue()

    ip_img = display.Image(data=data, format='png', embed=True)
    return ip_img._repr_png_()

# register display func with PNG formatter:
png_formatter = get_ipython().display_formatter.formatters['image/png']
png_formatter.for_type(Image.Image, display_pil_image)
```

```python
def add_coords_to_df(df, xcolumn='date', ycolumn='kw', x=date_to_coord, 
                     y=kw_to_coord, inplace=True):
    if not inplace:
        df = df.copy()
    df['xcoord'] = x(df[xcolumn])
    df['ycoord'] = y(df[ycolumn])
    return df
```

```python
def create_pixelplot(df, color_scale=grays):
    points_counts = Counter([(y, x) for ind, y, x in df[['ycoord', 'xcoord']].itertuples()]).items()
    max_count = max([count for coords, count in points_counts])
    
    max_y = max([coords[0] for coords, color in points_counts])
    max_x = max([coords[1] for coords, color in points_counts])
    points_color = [(coords, color_scale(count, max_count)) for coords, count in points_counts]


    #Image.fromarray expects shape (y, x, 3)
    plot_array = np.ndarray((max_y, max_x, 3))
    for coords, color in points_color:
        y, x = coords
        plot_array[y-1][x-1] = np.uint8([color, color, color])
    image = Image.fromarray(np.uint8(plot_array))
    return image
```

```python
%%capture a
smaller_than_1000 = all_pv[all_pv['kw'] <= 1000].copy()
add_coords_to_df(smaller_than_1000)
```

```python
def log_to_lin_scale(val, in_max, out_min, out_max):
    #only works for positive domains and values > 0
    log_max = math.log(in_max)
    log_val = math.log(val)
    ratio = log_val/log_max
    out_span = out_max - out_min
    return (ratio*out_span) + out_min
```

```python
from functools import partial
```

```python
grays = partial(log_to_lin_scale, out_min=50, out_max=255)
```

```python
i = create_pixelplot(smaller_than_1000)
```
